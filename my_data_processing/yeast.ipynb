```python
import pandas as pd
import numpy as np
```

```python
df.iloc[:,-14:].sum(axis=0).sort_values(ascending=False)
```

```python
colums = df.iloc[:,-14:].sum(axis=0).sort_values(ascending=False).keys()[:4]
```

```python
df = df.loc[df.loc[:,colums].sum(axis=1) != 0, :]
df
```

```python
X = np.array(df.iloc[:,:103])
X
```

```python
Y = np.array(df.loc[:,colums])
Y
```

```python
train_size = int(len(X) * 0.7)
X_train = X[:train_size] 
Y_train = Y[:train_size]
X_test = X[train_size:]
Y_test = Y[train_size:]
```

```python
np.save('./yeast_train_data.npy',X_train)
np.save('./yeast_train_label.npy',Y_train)
np.save('./yeast_test_data.npy',X_test)
np.save('./yeast_test_label.npy',Y_test)
print('Finish')
```
